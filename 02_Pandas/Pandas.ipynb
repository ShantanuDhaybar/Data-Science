```python
import numpy as np
import pandas as pd
```

```python
# Example : 1

df = pd.DataFrame({'Name' : pd.Series(['Pooja', 'Raj', 'Abhay', 'Neha', 'Nikita', 'Sam', 'John', 'Virat', 'Rohan', 'Ishan']),
                   'Grade' : pd.Series(['A', 'B', 'A', 'C', 'A', 'B', 'A', 'C', 'C', 'B']),
                   'Val1' : np.random.rand(10).round(2),
                   'val2' : np.random.randint(1,100,10)})


df
```

```python
# Display records of 'Pooja

df[df.Name == 'Pooja']
```

```python
# Display all with Grade 'C'

df[df.Grade == 'C']
```

```python
# Display all with val2 greater than 40

df[df.val2 > 40]
```

```python
# Display all records of 'B','C' with val2 > 25

df[(((df.Grade == 'C') |  (df.Grade =='B')) & (df.val2 > 25) )]               # logical operators : and, or     F  T =>  T


# isin()
df[df.Grade.isin(['B', 'C'])]
```

```python
# Display All with Grade 'A' having val1 less than 0.8.

df[(df.Grade == 'A' ) &  (df.Val1 < 0.8)]
```

```python
# for String : str.contains()

# Display all records having 'A' in names.

df[df.Name.str.contains('A')]
```

```python
# Display all records having 'a' in names.

df[df.Name.str.contains('a')]
```

##Statistics

```python
# Statistics :

# Mean
# Median
# Mode
# Standard Daviation
# Variance
# Range
# IQR Inter Quartile Range
# Freaquency  : 1. Relative Frequency     2. Cumulative Frequency
```

```python
# Example : 
s1 = pd.Series([4, 5, 6, 10, 8, 8, 1,1,2, 4, 6, 8])
```

```python
# Marks : 4, 5, 6, 10, 8, 8, 1,1,2, 4, 6, 8

# 1. Mean = (Sum / Count) = 63/12 = 5.25

# 2. Median = 1,1,2,4,4,   5,6   ,6,8,8,8,10 =  (5 + 6) /2 = = 11/2 = 5.5

# 3. Mode = 8

# 4. Range = High - Low = 10 - 1 = 9

# 5. Standard Deviation  / Variance

# X        X- Mean              Sq(X-Mean)
# 4     4 - 5.25 = -1.25            
# 5     5 - 5.25 = -0.25           6.25
# 6     6 - 5.25 =  1.25
# 10    10       =  3.75
# 8              =  2.75


# Inter Quartile Range :

# 1,1,2  |    4,4,5, |   6,6,8,  |   8,8,10
#     Q1 = 3       Q2=5.5       Q3=8

# Frequency :

# Cake     NO(Freq.)      Relative Freq.            Cumulative Frequency
#  1       5                5 / 50 =  0.1                0.1
#  2      17               17/50  =   0.34                 0.10 + 0.34= 0.44
#  3      21               21/50  =   .42                0.44 + 0.42 = 0.86
#  4       7                7/50 =  .14                  0.86 + 0.14 = 1
```
