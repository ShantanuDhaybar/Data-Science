### Series

```python
a = [10, 20, 30, 40, 50]
print(a, type(a))
```

```python
arr1 = nm.array(a)
print(arr1, type(arr1))
```

```python
s1 = pd.Series(a)
print(s1, type(s1))
```

```python
# In pandas default index of Series starts from 0.
# In Series we can change index values.
```

```python
# changing Index
```

```python
s2 = pd.Series(a,index = ['a', 'b', 'c', 'd', 'e'] )

print(s2)
```

```python
s3 = pd.Series(a, index = [0,2,4,6,8])

print(s3)
```

```python
## Accessing Elements : index

# s1 : Display 30, 50

s1[2], s1[4]
```

```python
# s2 : 30, 50

print(s2[2], s2[4])
print(s2['c'], s2['e'])

# Elements can be accessed with both default and new indices.
```

```python
# s3 : 30, 50

#print(s3[1], s3[3])
print(s3[4], s3[8])

# In case of conflicting indices new index values will prevail.
```

```python
# 2. Slicing : series_name[start  : end : steps ]
```

```python
# Display 10,20,30

print(s1[0 :3 ])
```

```python
# Display 40,50

print(s1[3:5])
```

```python
# Display 10,30,50

print(s1[ : : 2])
```

```python
# Create Marks of 20 students out of 100 and convert it into series. [40 to 98]

m = nm.random.randint(40, 98, 20)
print(m, type(m))
```

```python
s5 = pd.Series(m)
print(s5)
```

```python
# Slicing :

# Display marks of first 5 students.

print(s5[ :5 ])
```

```python
# Display marks of last 5 students
print(s5[-5 :])
```

```python
# Display marks of alternate students.
print(s5[ : : 2])
```

```python
# Display marks of every 3rd student from 5 to 15.
print(s5[ 4: 15: 3])
```

```python
# sum, min, max, mean, median

print(" Sum of all Marks ", s5.sum())
print(" Min Value of all Marks ", s5.min())
print(" Max Value of all Marks ", s5.max())
print(" Mean of  Marks ", s5.mean())
print(" Median o Marks ", s5.median())
```
